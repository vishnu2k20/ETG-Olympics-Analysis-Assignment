# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("summer.csv")
sns.set_style("white")
sns.set_style("ticks")
```

## 1. In how many cities Summer Olympics is held so far?

```python
len(df["City"].unique())
```

## 2. Which sport is having most number of Gold Medals so far?

```python
df[df["Medal"]=="Gold"]["Sport"].value_counts().head().plot.bar(figsize = (10,6),color="blue")
plt.xticks(rotation=0, horizontalalignment="center",size=15)
plt.yticks(size=15)
plt.title("Sport having Most number of Gold Medals",size=18)
plt.xlabel("Name of the Sport",size=18)
plt.ylabel("Number of Medals",size=18)
df[df["Medal"]=="Gold"]["Sport"].value_counts().head()
```

## 3. Which sport is having most number of medals so far?

```python
df["Sport"].value_counts().head().plot.bar(figsize = (10,7),color="green")
plt.xticks(rotation=0, horizontalalignment="center",size=15)
plt.yticks(size=15)
plt.title("Sport having Most number of Medals so far",size=18)
plt.xlabel("Sport Name",size=18)
plt.ylabel("Number of Medals",size=18)
df["Sport"].value_counts().head()
```

## 4. Which player has won most number of medals?

```python
df["Athlete"].value_counts().head().plot.bar(figsize = (10,6),color="grey")
plt.xticks(rotation=15, horizontalalignment="center",size=13)
plt.yticks(np.arange(0,24,step=2),size=15)
plt.title("Player has won Most number of Medals",size=18)
plt.xlabel("Player Name",size=18)
plt.ylabel("Number of Medals",size=18)
df["Athlete"].value_counts().head()
```

## 5. Which player has won most number Gold Medals of medals?

```python
df[df["Medal"]=="Gold"]["Athlete"].value_counts().head().plot.bar(figsize = (13,8),color="red")
plt.xticks(rotation=0, horizontalalignment="center",size=13)
plt.yticks(np.arange(0, 19, step=2),size=15)
plt.title("Player won Most number of Gold Medals",size=20)
plt.xlabel("\nPlayer Name",size=20)
plt.ylabel("Number of Medals",size=20)
df[df["Medal"]=="Gold"]["Athlete"].value_counts().head()
```

## 6. In which year India won first Gold Medal in Summer Olympics?

```python
print(*df[(df["Medal"]=="Gold") &(df["Country"]=="IND")]["Year"].head(1).tolist())
```

## 7. Which event is most popular in terms on number of players? (Top 5)

```python
df["Event"].value_counts().head().plot.bar(figsize = (10,6),color="brown")
plt.xticks(rotation=0, horizontalalignment="center",size=15)
plt.yticks(size=15)
plt.title("Most Popular Event in terms of number of plyers",size=18)
plt.xlabel("Event Name",size=18)
plt.ylabel("Number of Players",size=18)
df["Event"].value_counts().head()
```

## 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df[(df["Medal"]=="Gold")&(df["Gender"]=="Women")]["Sport"].value_counts().head().plot.bar(figsize = (10,6),color="indigo")
plt.xticks(rotation=0, horizontalalignment="center",size=15)
plt.yticks(size=15)
plt.title("Sport having most Female Gold Medalists",size=18)
plt.xlabel("Sport Name",size=18)
plt.ylabel("Number of Medals",size=18)
df[(df["Medal"]=="Gold")&(df["Gender"]=="Women")]["Sport"].value_counts().head()
```
